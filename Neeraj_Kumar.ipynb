### Now we have less number of columns out of which some correspond to the post approval of loan
We are focusing on analyzing borrower details and identifying the key factors contributing to loan defaults before approving loans.
Therefore, it is appropriate to remove columns or variables that don't pertain to this scenario.
Some columns like "id", "member_id", "url", "title", "emp_title", "zip_code", "last_credit_pull_d", and "addr_state" don't provide meaningful information for assessing loan defaults, so they can be excluded.
The "desc" column contains descriptive text data, which is currently not useful for our analysis, so we will also remove it.
Columns such as "out_prncp_inv" and "total_pymnt_inv" are more relevant to investors and don't contribute to analyzing defaults, so we will exclude them as well.
Since we already have "funded_amnt_inv," which shows the actual amount funded, the "funded_amnt" column is redundant and can be removed.

### The objective of this analysis is to identify individuals who are likely to default, which can only be determined for loans that are charged off.
###### Since no conclusions can be drawn for loans that are still active, we will exclude these records by removing loans with a "current" status.

````python
loan_data = loan_data[loan_data.loan_status != "Current"]
loan_data.loan_status.unique()
````

## Handling Missing values
 - columns with missing values are "emp_length", "revol_util".
 - So before doing that, lets see what kind of data each column has.

````python
loan_data.info()
````

````python
print("Mode : " + loan_data.emp_length.mode()[0])
loan_data.emp_length.value_counts()
````

### The value counts above indicate that the mode occurs significantly more frequently than the next most common value.
###### This suggests that we can confidently impute the null values in the column with the mode.
###### Additionally, since the percentage of missing values is very low, filling them with the mode will have minimal impact on the overall analysis.

````python
loan_data.emp_length.fillna(loan_data.emp_length.mode()[0], inplace = True)
loan_data.emp_length.isna().sum()
````

````python
loan_data.dropna(axis = 0, subset = ['revol_util'] , inplace = True)
loan_data.revol_util.isna().sum()
````

## Standardizing the data
###### The "revol_util" column, although categorized as an object column, contains continuous values. Therefore, we need to standardize the data in this column.
###### The "int_rate" column is another example of this. For the "emp_length" column, values of less than one year will be assumed as 0, while 10 or more years will be considered as 10.
###### Although the data type of "term" could be argued to be an integer, it only has two distinct values throughout the column, so it may be more appropriate to treat it as a categorical variable.

````python
loan_data.revol_util = pd.to_numeric(loan_data.revol_util.apply(lambda x : x.split('%')[0]))
````

````python
loan_data.int_rate = pd.to_numeric(loan_data.int_rate.apply(lambda x : x.split('%')[0]))
````

````python
loan_data.emp_length = pd.to_numeric(loan_data.emp_length.apply(lambda x: 0 if "<" in x else (x.split('+')[0] if "+" in x else x.split()[0])))
````

## Outlier Treatment

````python
sns.boxplot(loan_data['annual_inc'])
````

#### The data clearly indicates the presence of outliers, so we will remove them.
###### Let's examine the quantile information to determine the appropriate action.
###### Values above the 95th percentile appear disconnected from the general distribution, with a sharp increase for small changes in quantiles.
###### Therefore, we will set the threshold for outlier removal at the 95th percentile.

````python
quantile_info = loan_data.annual_inc.quantile([0.5, 0.75,0.90, 0.95, 0.97,0.98, 0.99])
quantile_info
````

````python
per_95_annual_inc = loan_data['annual_inc'].quantile(0.95)
loan_data = loan_data[loan_data.annual_inc <= per_95_annual_inc]
````

# Observations / Insights Gained for past defaulting customers

The analysis of charged-off loans reveals a higher probability of default in the following cases:

- Applicants with an income between 60k and 70k who take out loans for "home improvement."
- Applicants with "MORTGAGE" as their home ownership status and an income of 60k-70k.
- Applicants earning 70k-80k who receive an interest rate of 21-24%.
- Applicants with loan amounts between 30k and 35k, charged an interest rate of 15-17.5%.
- Applicants taking loans for small businesses with amounts exceeding 14k.
- Applicants with "MORTGAGE" home ownership and loan amounts between 14k and 16k.
- Applicants with a grade of "F" and loan amounts ranging from 15k to 20k.
- Applicants with 10 years of employment and loan amounts between 12k and 14k.
- Verified loans with amounts over 16k.
- Applicants with a grade of "G" and interest rates above 20%.



## Recommendations for Lending Club company for better profitabilty by avoiding charging off loans:
### Screening House Ownership:
 Avoid high-risk applicants with house ownership status marked as 'RENT' or 'MORTGAGE' (especially when income is within the 60k-80k range), as these applicants have shown a higher likelihood of default.

### Loan Purpose & Debt Consolidation:
 Prioritize applicants with loan purposes other than debt consolidation, home improvement, or small business loans, as these purposes are associated with higher default rates.

### Interest Rate Caps:
Set a maximum interest rate cap below 13%, as applicants charged rates in the 13-24% range are more prone to defaulting.

### Income Range Filters:
Avoid applicants with an income range of 31,201 - 58,402 and 60k-80k, especially when paired with loans for home improvement or other non-essential uses, as these income brackets appear at higher risk.

### Loan Amount Control:
Be cautious with applicants seeking loans between $5,429 - $10,357 or $15k - $35k, particularly those with a DTI ratio between 12-18 and monthly installments in the range of 145 - 274.

### Employment Length Verification:
Ensure that applicants' employment length is verified and consider lengthier employment as a factor. 10 years of employment is a critical cutoff point, with higher risk associated with longer durations.

### Credit Enquiries and Records:
Avoid applicants with 0 inquiries in the last 6 months and no derogatory public records—counterintuitively, this indicates a higher likelihood of defaulting. Verified records and inquiries are essential.

### Loan Term Limit:
Prefer shorter loan terms. Defaults are more common in 36-month terms.

### Lower Grade Applicants:
Be more cautious with applicants in Grade B, B5, F, and G.
Defaults increase when loans are issued to these grades, especially with high loan amounts or interest rates exceeding 20% for Grade G applicants.
